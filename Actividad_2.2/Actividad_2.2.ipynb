```python
#Primero tenemos que mandar a traer las librerias que ocuparemos para manipular los datos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Para poder manipular los datos, tenemos que hacer que python lea el CSV (reutlize la instrucción de la anterior actividad)
df= pd.read_csv('Boston_United States.csv') 
df
```

```python
#Primero quitaremos el signo de porcentaje de la columna del host_acceptance_rate (no sé porque no es la columna de response_rate la que contenga el dato requerido)
df['host_acceptance_rate']=df['host_acceptance_rate'].str.replace('%', '')
df
```

```python
#Y despues convertimos el tipo de dato a str
df['host_acceptance_rate']= df['host_acceptance_rate'].astype(float)
df
```

```python
#Corremos un .info para comprobar que los datos si están en tipo float ya que así ahora se podrá manipular como un número y no como una cadena
df.info()
```

```python
#Una vez que tenemos listos los datos del acceptance_rate, podemos pasar a filtrarlos por aquellos que sean iguales o mayores a 95 y que no se califiquen como superhost (o sea el atributo f)
df1=df[(df['host_acceptance_rate'] >= 95) & (df['host_is_superhost'] == "f")]
df1
```

```python
#Una vez listo el filtro, solo lo pasamos a un csv
df1.to_csv("Primer_punto.csv")
```

```python
#Como segundo filtro aplicamos la condicional de negación y la condición que negará serán las filas que contengan un valor igual a 1 para la columna de baños
df2 = df[~(df['bathrooms'] == 1.0)]
df2.to_csv("Segundo_punto.csv")
```

```python
#Para el siguiente filtro necesitamos manipular la columna de price, sin embargo, el tener los carácteres: "$", ",", necesitamos quitarlos
df['price']=df['price'].str.replace('$', '')
df['price']=df['price'].str.replace(',', '')
df
```

```python
#Una vez que quitamos los caráceteres, ahora pasamos el dato a tipo float para que finalmente el programa lo reconozca como número y pueda aplicarse el filtro
df['price']= df['price'].astype(float)
df
```

```python
#Comprobamos el tipo de dato de la columa "price" y que este sea tipo float
df.info()
```

```python
#Una vez listo, ahora aplicamos un filtro que solo deje a los precios que sean iguales o mayores a 100, pero que sean menores o iguales a 200
df3 = df[(df['price'] >= 100) & (df['price'] <= 200)]
df3
```

```python
#Una vez listo el filtro, solo lo pasamos a un csv
df3.to_csv("Tercer_punto.csv")
```

```python
#Dado a que no elegí la variable de "instant_bookable" dentro de mi base lo reemplaze por la más similar que era "has_availability", y se aplica el filtro con el carácter "t", así lo mimso en el caso de la segunda variable. Y aplicamos el filtro
df4 = df[(df['has_availability'] == 't') & (df['host_identity_verified'] == 't')]
df4
```

```python
#Una vez listo el filtro, solo lo pasamos a un csv
df4.to_csv("Cuarto_punto.csv")
```

```python
#Aquí solo aplicamos la condicional en la columna de room_type para los tipos de habitación que nos requiere la actividad y utilizamos como conector "or0"
df5 = df[(df['room_type']=='Hotel room') | (df['room_type'] == 'Private room')]
df5
```

```python
#Una vez listo el filtro, solo lo pasamos a un csv
df5.to_csv("Quinto_punto.csv")
```

```python
#Y en este caso solo filtramos la columna de amenidades a aquellas que sean posean más de 4
df6 = df[df['accommodates'] > 4]
df6
```

```python
#Una vez listo el filtro, solo lo pasamos a un csv
df6.to_csv("Sexto_punto.csv")
```
