```python
#Para poder manipular los datos, tenemos que hacer que python lea el CSV (sé que suena a IA la instrucción pero le prometo que solo así me explico)
df= pd.read_csv('Boston_United States.csv') 
df
```

```python
#Ya que como no elegí el nombre de los hos como variable, opte por elegir el vecindario como filtro, pero a la hora de intentar hacer el filtro
#como intente hacer la comparativa pero mostraba un error de longitud de datos, opte por crear data frames con el atributo de la selección de nombres
df1 = [df[df["host_neighbourhood"] == "Back Bay"],
      df[df["host_neighbourhood"] == "Dorchester"],
      df[df["host_neighbourhood"] == "Roxbury"],
      df[df["host_neighbourhood"] == "Jamaica Plain"],
      df[df["host_neighbourhood"] == "North End"]]

df1_listo = pd.concat(df1) #Uní todos los frames con un .concat


```

```python
#Una vez listo ya solo lo exporte en forma de csv
df1_listo.to_csv("Primer_punto.csv", index=False)
```

```python
#Aquí cree el segundo filtro en el que la columna de host_since solo dejará los valores que correspondan despues del primero de enero de 2020, para así saber de los host registrados desde 2020 para adelante
df2 = df[df['host_since'] > '2020-01-01']
df2
```

```python
#Una vez listo ya solo lo exporte en forma de csv
df2.to_csv("Punto 2.csv")
```

```python
#Primero cree un "nuevo data Frame" ya que quería hacer manipulable la columna del host response time, ya que queríe elegir todos los tiempos que correspondieran o enteraran a menos de un día de respuesta
df1= pd.read_csv('Boston_United States.csv',index_col=4)
df1
```

```python
#Aquí hice el filtro un avez listos los datos para ser manipulados e incluí las variables que antes mencioné
df3= df1.loc[['within a day', 'within a few hours', 'within an hour']]
df3
```

```python
#Una vez listo ya solo lo exporte en forma de csv
df3.to_csv("Tercer_punto.csv")
```

```python
#Alguien en la clase comentó que lo hizon los residuos así que robe el tip y solo puse que el residuo fuera igual a 0 si se dividía entre 0 y le indique que los número que dicidiría serían los indíces
df4 = df[df.index % 200 == 0] 
df4
```

```python
#Una vez listo ya solo lo exporte en forma de csv
df4.to_csv("Cuarto_punto.csv")
```

```python
#Ahora que lo analizo así pude resolver dos ejercicio anteriores, pero, aquí utilice el conector "or" para indicarle que solo filtrara por aquellas que fueran entire o en su caso, private room
df5 = df[(df['room_type']=='Entire home/apt') | (df['room_type'] == 'Private room')]
df5
```

```python
#Una vez listo ya solo lo exporte en forma de csv
df5.to_csv("Quinto_punto.csv")
```

```python
#Aquí solo filtramos la columna denomidada "beds", e incluimos todos los valores que sean menores o iguales a 4
df6 = df[df['beds'] <= 4]
df6
```

```python
#Una vez listo ya solo lo exporte en forma de csv
df6.to_csv("Sexto_punto.csv")
```

```python
#La columna que se filtrará será la que comprueba si es un superhost, y aquí como la categorización de los valores solo se da por true o false (abreviado), solo indicamos que seleccione los valores verdaderos o sea "t"
df7 = df[df['host_is_superhost']=='t']
df7
```

```python
#Una vez listo ya solo lo exporte en forma de csv
df7.to_csv("Septimo_punto.csv")
```

```python
#Aquí como el filtro que aplicaremos será por la posición de los indices usamos el .iloc, y dentro de los parentesis indicaremos "las instrcciones" en las que el programa filtrará los datos, le decimos que empezará en la posición 1, no tendrá un límite donde paré, e irá en saltos de dos posiciones
df8 = df.iloc[1::2]
df8
```

```python
#Una vez listo ya solo lo exporte en forma de csv
df8.to_csv("Octavo_punto.csv")
```
